```python
import whisper

def transcribe_video(video_path, output_txt):
    # Выбор модели: 'base', 'small', 'medium', 'large'
    # Для русского языка рекомендую 'small' (быстро) или 'medium' (качественно)
    model = whisper.load_model("small") 

    print(f"Начинаю обработку: {video_path}")
    
    # transcribe автоматически извлечет аудио из mp4
    # language='ru' помогает модели не путать языки в начале
    result = model.transcribe(video_path, language='ru', verbose=False)

    # Сохраняем текст в файл
    with open(output_txt, "w", encoding="utf-8") as f:
        f.write(result["text"])
    
    print(f"Готово! Текст сохранен в {output_txt}")

# Использование
transcribe_video(
    "ArchiveNTV/2003/1_01_01_2003_Karpova_Wasserman_Shtefan.mp4",
    "transcript_01_01_2003.txt"
)
```
